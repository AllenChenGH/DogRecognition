<h2>1. Preparation</h2>

```python
import pandas
import numpy
import pickle
import time

from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier

from variables import pickle_path, temp_path
```

```python
df_results = pandas.DataFrame(columns = ['race number', 'cluster number','classifier', 'accuracy', 'best parameters'])
```

```python
N_RACES = 10
N_CLUSTERS = 1000

# load data
df_model_train = pickle.load(open(pickle_path + "df_model_train_with_%s_clusters_%s_races.p"%(N_CLUSTERS, N_RACES), "rb" ))
df_model_test = pickle.load(open(pickle_path + "df_model_test_with_%s_clusters_%s_races.p"%(N_CLUSTERS,N_RACES), "rb" ))

cluster_columns = ["cluster_%s"%i for i in range(0, N_CLUSTERS)]

X_train = df_model_train[cluster_columns].values
X_test = df_model_test[cluster_columns].values

y_train = list(df_model_train['encoded_race'].values)
y_test = list(df_model_test['encoded_race'].values)
```

<h2>2. Logistic Regression</h2>

```python
# cross validation
cv_lr = model_selection.GridSearchCV(LogisticRegression(), {'penalty':['l1','l2'], 'C':[0.1,0.5,1.0], 'n_jobs':[4]}, cv=5, scoring='accuracy')

# fit
cv_lr.fit(X_train, y_train)

best_params_lr= cv_lr.best_params_
print("best parameters: %s"%best_params_lr)

# test with best parameters
final_model = LogisticRegression(penalty=best_params_lr['penalty'], C=best_params_lr['C'], n_jobs=4)
final_model.fit(X_train, y_train)
y_predict = final_model.predict(X_test)
final_accuracy_lr = metrics.accuracy_score(y_test, y_predict)

results = {
    'race number':N_RACES,
    'cluster number':N_CLUSTERS,
    'classifier':'Logistic Regression',
    'accuracy':final_accuracy_lr,
    'best parameters':best_params_lr
}
df_results = df_results.append(results, ignore_index=True)
```

<h2>3. Support Vector Classifier</h2>

```python
# cross validation
cv_svc = model_selection.GridSearchCV(LinearSVC(), {'C':[0.1,0.5,1.0]}, cv = 5, scoring = 'accuracy')

# fit
cv_svc.fit(X_train, y_train)

best_params_svc= cv_svc.best_params_
print("best parameters: %s"%best_params_svc)

# test with best parameters
final_model = LinearSVC(C = best_params_svc['C'])
final_model.fit(X_train, y_train)
y_predict = final_model.predict(X_test)
final_accuracy_svc = metrics.accuracy_score(y_test, y_predict)

results = {
    'race number':N_RACES,
    'cluster number':N_CLUSTERS,
    'classifier':'Support Vector Classifier',
    'accuracy':final_accuracy_svc,
    'best parameters':best_params_svc
}
df_results = df_results.append(results, ignore_index=True)
```
